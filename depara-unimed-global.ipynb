# Depara Unimed ↔ Global

**Objetivo:** mapear produtos entre o catálogo Unimed (`COD_PRODUTO`) e a Curva ABC Global (`Cod Item`).

## Resumo das bases

| | Unimed | Global (Curva ABC) |
|---|---|---|
| Granularidade | 1 linha = 1 movimentação de entrada | 1 linha = 1 SKU do CD |
| Produtos únicos | 3.067 | 2.038 |
| Faixa de código | 2 – 5.889 (mediana ~3.247) | 83.639 – 117.804 (mediana ~87.484) |
| Descrição | UPPERCASE, abreviada (~30 chars) | lowercase, detalhada (~42 chars) |
| Campos extras | MARCA, PRINCIPIO_ATIVO, LINHA_PRODUTO | ABC/PQR/XYZ, VL Médio, Previsão |

**Conclusão principal:** `COD_PRODUTO` e `Cod Item` são **sistemas de codificação distintos** — não há overlap numérico. O depara precisa ser feito por **descrição normalizada** (idealmente usando `LINHA_PRODUTO` da Unimed).

````python
# --- Checagem: COD_PRODUTO == Cod Item? ---
global_codes = set(global_df['COD_PRODUTO'])
unimed_codes = set(unimed_catalogo_df['Cod Item'])
overlap = global_codes & unimed_codes

print(f"Produtos únicos Global (dist.): {len(global_codes):,}")
print(f"Produtos únicos Unimed:         {len(unimed_codes):,}")
print(f"Códigos em comum:               {len(overlap):,}")
print(f"\nFaixa Global:  {min(global_codes)} – {max(global_codes)}")
print(f"Faixa Unimed:  {min(unimed_codes)} – {max(unimed_codes)}")
print(f"\n→ Os códigos NÃO batem. São identificadores de sistemas diferentes.")
````

````python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# 1) Distribuição ABC Global
unimed_catalogo_df['ABC'].value_counts().sort_index().plot.bar(ax=axes[0, 0], color=['#e74c3c', '#f39c12', '#3498db'])
axes[0, 0].set_title('Unimed — Curva ABC')
axes[0, 0].set_ylabel('Qtd produtos')

# 2) Top princípios ativos Unimed
top_pa = global_df.drop_duplicates('COD_PRODUTO')['PRINCIPIO_ATIVO'].value_counts().head(10)
top_pa.plot.barh(ax=axes[0, 1], color='#2ecc71')
axes[0, 1].set_title('Global — Top 10 Princípios Ativos')
axes[0, 1].invert_yaxis()

# 3) Comprimento das descrições
global_lens_desc = global_df.drop_duplicates('COD_PRODUTO')['DESCRICAO_PRODUTO'].str.len()
unimed_lens_desc = unimed_catalogo_df['Desc Item'].str.len()
axes[1, 0].hist(global_lens_desc['DESCRICAO_PRODUTO'].str.len(), bins=30, alpha=0.7, label='Unimed', color='#3498db')
axes[1, 0].hist(unimed_lens_desc.str.len(), bins=30, alpha=0.7, label='Global', color='#e74c3c')
axes[1, 0].set_title('Comprimento das descrições')
axes[1, 0].legend()

# 4) Exemplo: mesma LINHA_PRODUTO, códigos diferentes (marcas distintas)
exemplo = global_df[global_df['LINHA_PRODUTO'].str.contains('DEXAMETASONA', na=False)][
    ['COD_PRODUTO', 'DESCRICAO_PRODUTO', 'MARCA', 'LINHA_PRODUTO']
].drop_duplicates('COD_PRODUTO')
axes[1, 1].axis('off')
text = "Exemplo: mesma LINHA_PRODUTO, COD_PRODUTO diferente\n\n"
for _, r in exemplo.head(5).iterrows():
    text += f"  {r['COD_PRODUTO']:>5} | {r['MARCA']:<12} | {r['DESCRICAO_PRODUTO'][:40]}\n"
axes[1, 1].text(0.02, 0.95, text, transform=axes[1, 1].transAxes, fontsize=9,
                verticalalignment='top', fontfamily='monospace')

plt.tight_layout()
plt.show()
````

## Estratégia sugerida para o depara

Com base no fuzzy matching completo (3.067 × 2.038 comparações):

| Faixa de similaridade | Qtd Unimed | % | Ação sugerida |
|---|---|---|---|
| ≥ 0.9 | 301 | 9.8% | Match automático |
| 0.8 – 0.9 | 624 | 20.4% | Match automático com revisão amostral |
| 0.7 – 0.8 | 817 | 26.6% | Revisão manual |
| 0.5 – 0.7 | 1.235 | 40.3% | Revisão manual ou regras extras |
| < 0.5 | 99 | 3.2% | Sem match (majoritariamente insumos/cosméticos) |

**Pontos de atenção:**
1. **`LINHA_PRODUTO` é melhor que `DESCRICAO_PRODUTO`** em 1.371 casos — usar como campo primário de matching
2. **Múltiplos COD_PRODUTO → 1 Cod Item** é esperado (mesmo medicamento, marcas diferentes) — ex: 4 códigos de Dipirona 500mg/ml → Global 83668
3. **1.529 itens Global** (75%) não têm match Unimed com sim ≥ 0.7 — muitos são biológicos/oncológicos de alto valor (Dupixent, Stelara, etc.)
4. **278 descrições Unimed** têm >1 código (cateter, cloreto de sódio, agulhas) — depara deve ser N:1
5. **Princípio ativo não serve diretamente** — nomenclaturas incompatíveis entre bases (Unimed usa nome comercial/ANVISA, Global usa DCB lowercase)

## Granularidade dos produtos Unimed

Cada `COD_PRODUTO` é um **SKU comercial** (marca + apresentação). A apresentação clínica fica em `LINHA_PRODUTO`.

```
PRINCIPIO_ATIVO  (652)     → substância
    └── LINHA_PRODUTO (1.424) → dose + forma + via + volume   ← nível Global
            └── COD_PRODUTO (3.067) → + marca/laboratório
                    └── DESCRICAO_PRODUTO (2.678) → nome comercial (SANDOSTATIN, BUSCOPAN…)
```

| Nível | Campo | O que captura | Exemplo |
|---|---|---|---|
| 1 | `PRINCIPIO_ATIVO` | Substância | `DIPIRONA` |
| 2 | `LINHA_PRODUTO` | Apresentação clínica completa | `DIPIRONA (500MG/ML SOLUCAO INJETAVEL AMPOLA 2ML)` |
| 3 | `COD_PRODUTO` | SKU = linha + marca | `275` TEUTO, `3303` HYPOFARMA |
| — | `DESCRICAO_PRODUTO` | Nome comercial/abreviado | `DIPIRONA 500MG/ML INJ IV/IM 2ML VP` |

**Para o depara:** Global opera no **nível 2** (`LINHA_PRODUTO`). Vários `COD_PRODUTO` Unimed → 1 `Cod Item` Global.

````python
# DataFrame de produtos únicos — base para similaridade
global_produtos = global_df.drop_duplicates(subset=['COD_PRODUTO']).copy()

# Parse LINHA_PRODUTO → nome + detalhe clínico
import re
def parse_linha(linha):
    linha = str(linha).strip()
    m = re.match(r'^(.+?)\s*\((.+)\)\s*$', linha)
    return (m.group(1).strip(), m.group(2).strip()) if m else (linha, None)

parsed = unimed_produtos['LINHA_PRODUTO'].apply(parse_linha)
unimed_produtos['linha_nome'] = parsed.apply(lambda x: x[0])
unimed_produtos['linha_detalhe'] = parsed.apply(lambda x: x[1])

print("Granularidade:")
for col in ['PRINCIPIO_ATIVO', 'linha_nome', 'LINHA_PRODUTO', 'DESCRICAO_PRODUTO', 'COD_PRODUTO']:
    print(f"  {col:25s} → {unimed_produtos[col].nunique():,} únicos")

# COD por LINHA
cod_por_linha = unimed_produtos.groupby('LINHA_PRODUTO')['COD_PRODUTO'].nunique()
print(f"\nLINHA com 1 COD (marca única):  {(cod_por_linha == 1).sum():,} ({(cod_por_linha==1).mean()*100:.0f}%)")
print(f"LINHA com 2+ CODs (multi-marca): {(cod_por_linha > 1).sum():,} ({(cod_por_linha>1).mean()*100:.0f}%)")
print(f"Máximo CODs por LINHA:           {cod_por_linha.max()}")

# Exemplo visual
ex = unimed_produtos[unimed_produtos['LINHA_PRODUTO'].str.contains('DIPIRONA \\(500MG', na=False)]
ex[['COD_PRODUTO', 'MARCA', 'DESCRICAO_PRODUTO', 'LINHA_PRODUTO']].head(8)
````

## Fluxo em 2 fases

### Fase 1 — Depara (match clínico)
Unidade: `LINHA_PRODUTO` ↔ `Cod Item` Global  
Campo de similaridade: `LINHA_PRODUTO` (primário) + `DESCRICAO_PRODUTO` (secundário)

### Fase 2 — Análise de preços
Unidade: `COD_PRODUTO` + `MARCA` (dentro de cada match clínico)

| Fonte | Preço | Granularidade |
|---|---|---|
| Unimed | `CUSTO_ENTRADA` | por SKU/marca (`COD_PRODUTO`) |
| Global | `VL Médio (R$)` | por apresentação clínica (`Cod Item`) |

**Exemplo:** linha clínica `DIPIRONA (500MG/ML SOL INJ…)` → Global 83668 (R$ X)  
Depois, comparar preço Global vs cada marca Unimed:

```
Global 83668  VL Médio = R$ ?
  ├─ COD 275   TEUTO      CUSTO_ENTRADA = R$ ?
  ├─ COD 3303  HYPOFARMA  CUSTO_ENTRADA = R$ ?
  └─ COD 4784  SANTISA    CUSTO_ENTRADA = R$ ?
```

Global não tem marca → comparação é **apresentação clínica vs custo por laboratório**.

## Fase 1 — Comparação de métodos de similaridade

Unidade de match: **`LINHA_PRODUTO`** (1.424 linhas clínicas) ↔ **`Cod Item`** Global (2.038)

| Método | Biblioteca | Lógica |
|---|---|---|
| `fuzz_token_set` | thefuzz | tokens em comum, ignora ordem/extras |
| `fuzz_token_sort` | thefuzz | tokens ordenados |
| `fuzz_wratio` | thefuzz | weighted ratio (parcial) |
| `tfidf` | scikit-learn | char n-grams (3–5) + cosine |
| `spacy` | spaCy pt_core_news_md | word vectors + fallback jaccard de lemmas |

````python
from depara.fase1_similarity import run_all_methods, method_summary, method_agreement

matches_long, fase1 = run_all_methods(GLOBAL_DISTRIBUIDOR_PATH, UNIMED_CATALOGO_PATH)

print(f"Linhas clínicas Unimed: {len(fase1):,}")
method_summary(fase1).round(3)
````

````python
import matplotlib.pyplot as plt
import seaborn as sns

summary = method_summary(fase1)
agree = method_agreement(fase1)

fig, axes = plt.subplots(1, 3, figsize=(16, 4))

score_cols = ['fuzz_token_set', 'fuzz_token_sort', 'fuzz_wratio', 'tfidf', 'spacy']
fase1[score_cols].boxplot(ax=axes[0])
axes[0].set_title('Distribuição de scores')
axes[0].set_ylabel('similaridade')
axes[0].tick_params(axis='x', rotation=45)

axes[1].bar(summary['method'], summary['pct_ge_0.8'], color='#2ecc71')
axes[1].set_title('Match confiável (score ≥ 0.8)')
axes[1].tick_params(axis='x', rotation=45)

axes[2].barh(agree['method_a'] + ' × ' + agree['method_b'], agree['agreement_pct'])
axes[2].set_title(f'Concordância de cod_item\n(todos iguais: {agree.attrs["all_methods_same_pct"]:.1f}%)')
axes[2].set_xlabel('%')

plt.tight_layout()
plt.show()

print("\nConfiança do match recomendado:")
print(fase1['confianca'].value_counts().to_string())
````

````python
# Amostra para revisão manual — consenso vs discordância
cols_show = [
    'linha_produto', 'principio_ativo', 'n_skus',
    'fuzz_token_set', 'spacy', 'tfidf', 'score_mean', 'confianca',
    'best_cod_item', 'best_desc_global',
]

print("=== Alta confiança (amostra) ===")
display(fase1[fase1['confianca'] == 'alta'][cols_show].head(8))

print("\n=== Revisar manualmente (amostra) ===")
display(fase1[fase1['confianca'] == 'revisar'].nsmallest(8, 'score_mean')[cols_show])

print("\n=== Discordância entre métodos (amostra) ===")
discord = fase1.nlargest(8, 'score_std')[
    ['linha_produto', 'fuzz_token_set', 'fuzz_wratio', 'spacy', 'tfidf',
     'desc_fuzz_token_set', 'desc_spacy']
]
display(discord)
````

## Fase 1b — Match via LLM (pydantic-ai)

Pipeline híbrido: **fuzzy recupera top-K candidatos** → **LLM escolhe o match clínico**.

```
LINHA_PRODUTO → fuzzy top-12 → pydantic-ai Agent → cod_item Global
                                      ↓
                              cache SQLite (prod)
```

**Módulos:** `depara/llm/` (schemas, agent, matcher, cache) · **CLI:** `depara.cli`

```bash
# Teste local (sem API — valida pipeline)
python -m depara.cli test --limit 5

# Produção — prioriza linhas "revisar" da fase fuzzy
python -m depara.cli run --filter revisar

# Item específico
python -m depara.cli one "DIPIRONA (500MG"
```

Configure `.env` (copie de `.env.example`): `DEPARA_MODEL`, `OPENAI_API_KEY`.

````python
from depara.llm.config import DeparaLLMSettings
from depara.llm.matcher import LLMMatcher
from depara.llm.candidates import retrieve_candidates, format_candidates_prompt, format_unimed_prompt

# Preview: candidatos fuzzy para uma linha (sem chamar LLM)
settings = DeparaLLMSettings(mode="prod")
matcher = LLMMatcher(settings)
row = matcher.unimed_linhas.iloc[0]
item = matcher._row_to_input(row)
candidates = retrieve_candidates(item, matcher.global_items, top_k=8)

print(format_unimed_prompt(item))
print("\n--- Candidatos ---")
print(format_candidates_prompt(candidates))
````
